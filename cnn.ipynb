```python
train_datagen = ImageDataGenerator( 
        rescale=1./255,
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
        

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    "data/train",
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
    "data/val",
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical')
```

```python
!ls 
```

```python
modele = Sequential()
modele.add(Conv2D(32, (3, 3), activation='relu',
input_shape=(150, 150, 3)))
modele.add(MaxPooling2D((2, 2)))

modele.add(Conv2D(64, (3, 3), activation='relu'))
modele.add(MaxPooling2D((2, 2)))

modele.add(Conv2D(128, (3, 3), activation='relu'))
modele.add(MaxPooling2D((2, 2)))

modele.add(Conv2D(128, (3, 3), activation='relu'))
modele.add(MaxPooling2D((2, 2)))
modele.add(Flatten())

modele.add(Dropout(0.5))
modele.add(Dense(128, activation='relu'))
modele.add(Dense(36, activation='softmax'))

modele.compile(loss='categorical_crossentropy',
optimizer= 'Adam',
metrics=['acc'])
```

```python
modele.fit_generator(
    train_generator,
    steps_per_epoch=50,
    epochs=100,
    validation_data=validation_generator,
    validation_steps=300)
```

```python
modele.save('digits.h5')
```
